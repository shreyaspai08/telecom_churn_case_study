```python
#Analysing null values
null_perc = 100*(df.isnull().sum()/len(df.index))
null_perc
```

```python
def get_cols_with_missing_data_for_month(month):
    cols_with_missing_data = []
    for item in null_perc.items():
        if item[1]>0.0 and item[1]<10.0:
            if item[0].endswith('_'+str(month)):
                cols_with_missing_data.append(item[0])
    return cols_with_missing_data
```
